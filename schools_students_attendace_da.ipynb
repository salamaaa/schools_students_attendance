```python
from sqlalchemy import create_engine
import urllib.parse

#Step1: Connect to PostrgreSQL

username = "postgres"
password = urllib.parse.quote_plus("123Aa@456") #encode special characters '@' of the password
host = "localhost"
port = 5433
database = "schools_students_attendance_db"

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

#Step2: Load DataFrame into PostgreSQL
table_name = "attendance"
df.to_sql(table_name,engine,if_exists="replace",index=False)
print(f"Data Successfull loaded into table '{table_name}' in database '{database}'")
```
